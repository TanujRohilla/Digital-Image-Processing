```python
from PIL import Image 
img = Image.open('cat4.jpg')
pixels = img.load()
display(img)
m = []
for i in range(img.size[0]):
    row = []
    for j in range(img.size[1]):
        row.append(pixels[i, j][0])
    m.append(row)
fx = [
    [-1,-2,-1],
    [0,0,0],
    [1,2,1]
]
fy = [
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
]
def getSum(sobel_operator, grid):
    sum = 0;
    for i in range(len(grid)):
        for j in range(len(grid[0])):
              sum += sobel_operator[i][j] * grid[i][j]
    return sum

for i in range(1, img.size[0]-1):
    for j in range(1, img.size[1]-1):
        grid = [
            [m[i-1][j-1],  m[i-1][j], m[i-1][j+1]],
            [m[i][j-1],    m[i][j],   m[i][j+1]],
            [m[i+1][j-1],  m[i+1][j], m[i+1][j+1]]
           ]
        x = getSum(fx, grid)
        pixels[i, j] = (int(x), int(x), int(x))

print("After sobel operation on x axis")
display(img)

m = []
for i in range(img.size[0]):
    row = []
    for j in range(img.size[1]):
        row.append(pixels[i, j][0])
    m.append(row)

for i in range(1, img.size[0]-1):
    for j in range(1, img.size[1]-1):
        grid = [
            [m[i-1][j-1],  m[i-1][j], m[i-1][j+1]],
            [m[i][j-1],    m[i][j],   m[i][j+1]],
            [m[i+1][j-1],  m[i+1][j], m[i+1][j+1]]
           ]
        x = getSum(fy, grid)
        pixels[i, j] = (int(x), int(x), int(x))
print("Sobel operator along Y axis")
display(img)
```
